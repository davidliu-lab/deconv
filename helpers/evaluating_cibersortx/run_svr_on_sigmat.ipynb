```python
uri_sigmat = (
    "/Users/william/Downloads/liulab_mirror/data/evaluating_pseudobulks/"
    "csx_runs/pseudobulks/n_cells=5/malignant_from_one_sample=True/out/"
    "CIBERSORTx_inputrefscrnaseq_inferred_phenoclasses.CIBERSORTx_inputrefscrnaseq_inferred_refsample.bm.K999.txt"
)

df_sigmat = (
    pd.read_csv(uri_sigmat, sep="\t")
    .rename(columns={"NAME": "gene_symbol"})
    .set_index("gene_symbol")
)

df_sigmat
```

```python
uri_bulk_rnaseq_pseudobulks = (
    "/Users/william/Downloads/liulab_mirror/data/evaluating_pseudobulks/"
    "csx_input_files/"
    "bulk_rnaseq_pseudobulk.tsv"
)

df_bulk_rnaseq_pseudobulks = (
    pd.read_csv(
        uri_bulk_rnaseq_pseudobulks,
        sep="\t",
    )
    .rename(columns={"GeneSymbol": "gene_symbol"})
    .set_index("gene_symbol")
)

df_bulk_rnaseq_pseudobulks
```

```python
from sklearn.svm import NuSVR
```

```python
def fit_nu_svr(df_sigmat, df_bulk_rnaseq):
    nu_svr = NuSVR(nu=0.5, kernel="linear")
    genes_in_both = df_sigmat.index.intersection(df_bulk_rnaseq.index)
    X = df_sigmat.loc[genes_in_both]
    y = df_bulk_rnaseq.loc[genes_in_both].iloc[:, 0]
    nu_svr.fit(X, y)
    return nu_svr


nu_svr = fit_nu_svr(df_sigmat, df_bulk_rnaseq_pseudobulks)

nu_svr.coef_
```

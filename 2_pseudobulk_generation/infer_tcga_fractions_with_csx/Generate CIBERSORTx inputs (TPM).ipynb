```python
df_scrnaseq, df_metadata = helpers.datasets.load_jerby_arnon()
```

```python
single_cells_of_known_cell_type = df_metadata["cell_type"].dropna().sort_values().index
df_refsample = df_scrnaseq[single_cells_of_known_cell_type]
df_refsample = df_refsample.rename_axis(index="Gene", columns=None)


def look_up_cell_type(single_cell_id):
    return df_metadata.loc[single_cell_id][helpers.columns.CELL_TYPE]


df_refsample = df_refsample.rename(columns=look_up_cell_type)
df_refsample = df_refsample.reset_index()
```

```python
uri_refsample = "gs://liulab/tmp/refsample_jerby_arnon.txt"

df_refsample.to_csv(uri_refsample, sep="\t", index=False)
```

```python
uri_firebrowse_tcga = "gs://liulab/firebrowse.org/SKCM.rnaseqv2__illuminahiseq_rnaseqv2__unc_edu__Level_3__RSEM_genes__data.data.txt"

pd.read_csv(uri_firebrowse_tcga, sep="\t", header=[0, 1])
```

```python
df_tcga_skcm_hg19_scaled_estimate = (
    helpers.datasets.load_tcga_skcm_hg19_scaled_estimate_firebrowse()
)
```

```python
df_tcga_skcm_hg19_scaled_estimate
```

```python
df_mixtures = df_tcga_skcm_hg19_scaled_estimate.rename_axis(index="Gene", columns=None)
df_mixtures *= 1e6 / df_mixtures.sum()
df_mixtures = df_mixtures.reset_index()
```

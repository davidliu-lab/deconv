```python
gse115978_cell_annotations[["cells", "cell.types"]].set_index("cells")
```

```python
gse115978_tpm_formatted = (
    pd.merge(
        left=gse115978_tpm.iloc[:100].T,
        right=gse115978_cell_annotations[["cells", "cell.types"]].set_index("cells"),
        how="inner",
        left_index=True,
        right_index=True,
    )
    .set_index("cell.types")
    .rename_axis(None)
    .T.rename_axis("Gene")
)
# gse115978_tpm_formatted = gse115978_tpm_formatted.T
```

```python
gse115978_tpm_formatted
```

```python
out_path = "/mnt/buckets/liulab/william/gse115978_tpm_formatted.tsv"

gse115978_tpm_formatted.to_csv(
    out_path,
    sep="\t",
)
```

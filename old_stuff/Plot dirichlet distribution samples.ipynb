```python
import numpy as np
```

```python
np.set_printoptions(precision=8, suppress=True)
```

```python
random = np.random.default_rng(0)
```

```python
alpha = np.ones(shape=(10,)) / 10
alpha
```

```python
sample = random.dirichlet(alpha)
sample = np.sort(sample)[::-1]
sample
```

```python
import plotly.express as px
```

```python
fig = px.bar(sample)
fig.show("png")
```
